# FEATURE ENGINEERING

DROPPING COLUMNS CONTAINING SENTIMENT SCORE OF HEADLINE AND HEADLINE TEXT SINCE THEY WON'T AFFECT THE SENTIMENT SCORE OF TITLE

```python
dft1.drop(['Headline','SentimentHeadline'],axis=1,inplace=True)
```

```python
t=pd.get_dummies(dft1['Topic'],drop_first=True)
dft1.join(t)
dft1.drop('Topic',axis=1,inplace=True)
```

CREATING A SIMPLE BAG OF WORDS MODEL FOR THE TITLE COLUMN AND TAKING JUST 500 MAXIMUM WORDS TO AVOID HAVING TOO MANY FEATURES

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(max_features=500, binary=True,stop_words='english')
ti = vect.fit_transform(dft1.Title)

ti.toarray()
```

```python
df2 = pd.DataFrame(ti.toarray(), columns=vect.get_feature_names())
```

```python
df2.head()
```

```python
dft1.drop('Title',axis=1,inplace=True)
```

```python
dft1['Source'].isnull().sum()
```

FILLING THE MISSING VALUES IN SOURCE COLUMN WITH THE MODE 

```python
dft1['Source'].value_counts().head(100)
```

```python
dft1['Source'].fillna('Bloomberg',inplace=True)
```

SINCE THE SOURCE COLUMN CONTAINS MANY CATEGORIES , DOING ONE HOT ENCODING OR MAKING DUMMY VARIABLES MAY NOT BE THE BEST OPTION 

HENCE WE DID ONE HOT ENCODING FOR ONLY THE TOP 100 MOST OFTEN OCCURING CATEGORIES

```python
lst=dft1.Source.value_counts().sort_values(ascending=False).head(100).index
lst=list(lst)
```

```python
for categories in lst:
    dft1[categories]=np.where(dft1['Source']==categories,1,0)
```

```python
dft1.head()
```

```python
dft1.drop('Source',axis=1,inplace=True)
```

```python
dft1.head()
```

```python
dft1=dft1.join(df2)
```

```python
dft1.head()
```

# FITTING MODELS USING DIFFERENT REGRESSION ALGORITHMS

TO PREDICT SENTIMENT SCORE OF TITLE

```python
X=dft1.drop('SentimentTitle',axis=1)
Y=dft1['SentimentTitle']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, accuracy_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
import xgboost as xgb
from sklearn.svm import SVR
```

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train,y_train)
y_pred_lr = linear_regressor.predict(X_test)
print("---- Linear Regression - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_lr)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_lr)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_lr))))
```

```python
decision_regressor = DecisionTreeRegressor()
decision_regressor.fit(X_train,y_train)
y_pred_dt = decision_regressor.predict(X_test)
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_dt)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_dt)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_dt))))
```

```python
random_regressor = RandomForestRegressor(n_estimators=10)
random_regressor.fit(X_train,y_train)
y_pred_rf = random_regressor.predict(X_test)
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_rf)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_rf)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_rf))))
```

```python
adb_regressor = AdaBoostRegressor(base_estimator=linear_regressor, n_estimators=10)
adb_regressor.fit(X_train, y_train)
y_pred_adb = adb_regressor.predict(X_test)
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred_adb)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred_adb)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred_adb))))
```

```python
xg=xgb.XGBRegressor(random_state=42)
xg.fit(X_train, y_train)
y_pred = xg.predict(X_test)
```

```python
print("Mean Absolute Error (MAE): {}".format(mae(y_test, y_pred)))
print("Mean Squared Error (MSE): {}".format(mse(y_test, y_pred)))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(y_test, y_pred))))
```

HYPER PARAMETER TUNING FOR RANDOM FOREST ALGORITHM

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [10,30,50]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2,4,8,10]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

FOR SENTIMENT HEADLINE PREDICTION

```python
rf = RandomForestRegressor()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 33, cv = 3, verbose=2, random_state=42, n_jobs = -1)
rf_random.fit(X1, Y1)
```
